# W6Q1/Q2 — Email Campaign Performance (Onboarding + A/B Test)
**Question:** How does each email campaign perform across the full email funnel? Are there A/B test pairs? Which campaigns are underperforming?

**Audience:** Marketing Team  
**Data source:** `ANALYTICS.MARTS.MART_EMAIL_ENGAGEMENT`  
**SQL:** `sql/W6Q1Q2_email_campaign_performance.sql`

---

**Email funnel metrics:**
| Metric | Definition |
|--------|-----------|
| Delivery rate | delivered / sent |
| Open rate | opened / delivered |
| Click-to-open rate | clicked / opened |
| Bounce rate | bounced / sent |
| Spam rate | spam / sent |

## Setup

```python
import sys
sys.path.append('..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
from scipy import stats

from src.connection import query

sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams.update({'figure.dpi': 130})
PALETTE = sns.color_palette('muted', 8)
```

## Load Data

```python
with open('../sql/W6Q1Q2_email_campaign_performance.sql') as f:
    sql = f.read()

df = query(sql)
df['campaign_start'] = pd.to_datetime(df['campaign_start'])
df['campaign_end']   = pd.to_datetime(df['campaign_end'])
print(f"Campaigns: {len(df)}")
display(df)
```

## Engagement Rates Heatmap

```python
rate_cols = ['delivery_rate', 'open_rate', 'click_to_open_rate', 'bounce_rate', 'spam_rate']
rate_labels = ['Delivery Rate', 'Open Rate', 'Click-to-Open Rate', 'Bounce Rate', 'Spam Rate']

heatmap_data = df.set_index('email_campaign')[rate_cols].rename(columns=dict(zip(rate_cols, rate_labels)))

fig, ax = plt.subplots(figsize=(12, max(4, len(df) * 0.6)))
sns.heatmap(heatmap_data, annot=True, fmt='.1f', cmap='YlOrRd', ax=ax, linewidths=0.5)
ax.set_title('Email Engagement Rates by Campaign (%)', fontsize=13, pad=12)
ax.set_ylabel('')
plt.tight_layout()
plt.savefig('../output/W6Q1Q2_email_rates_heatmap.png', bbox_inches='tight')
plt.show()
```

## Volume Comparison

```python
vol_cols = ['total_sent', 'total_delivered', 'total_opened', 'total_clicked']
vol_labels = ['Sent', 'Delivered', 'Opened', 'Clicked']

plot_df = df[['email_campaign'] + vol_cols].rename(columns=dict(zip(vol_cols, vol_labels)))
plot_df = plot_df.melt(id_vars='email_campaign', var_name='Stage', value_name='Count')

fig, ax = plt.subplots(figsize=(14, 5))
sns.barplot(data=plot_df, x='email_campaign', y='Count', hue='Stage', palette='muted', ax=ax)
ax.set_title('Email Volume by Campaign and Stage', fontsize=13, pad=12)
ax.set_xlabel('')
ax.set_xticklabels(ax.get_xticklabels(), rotation=20, ha='right')
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
ax.legend(title='Stage')
sns.despine()
plt.tight_layout()
plt.savefig('../output/W6Q1Q2_email_volume_by_campaign.png', bbox_inches='tight')
plt.show()
```
