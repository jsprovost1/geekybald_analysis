# W4Q4 — Webpage Conversion Rate
**Question:** Which landing pages (first page visited) produce the highest trial conversion rates? Where should product and marketing focus to improve top-of-funnel?

**Audience:** Product Team  
**Data source:** `ANALYTICS.MARTS.MART_WEB_BEHAVIOR`  
**SQL:** `sql/W4Q4_webpage_conversion_rate.sql`

---

**Conversion definition:** `converted_to_trial = TRUE` (user went on to start a free trial).  
**Attribution:** Users are attributed to the first page they visited.

## Setup

```python
import sys
sys.path.append('..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
from scipy import stats

from src.connection import query

sns.set_theme(style='whitegrid', palette='muted')
plt.rcParams.update({'figure.dpi': 130})
PALETTE = sns.color_palette('muted', 8)
```

## Load Data

```python
with open('../sql/W4Q4_webpage_conversion_rate.sql') as f:
    sql = f.read()

df = query(sql)
print(f"Pages: {len(df)}")
display(df.sort_values('trial_conversion_rate', ascending=False))
```

## Conversion Rate by Landing Page

```python
df_sorted = df.sort_values('unique_visitors', ascending=False)

fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Volume
axes[0].barh(df_sorted['first_visit_page'], df_sorted['unique_visitors'], color=PALETTE[0])
axes[0].set_title('Unique Visitors by First Page', fontsize=12)
axes[0].set_xlabel('Unique Visitors')
sns.despine(ax=axes[0])

# Conversion rate
df_rate = df.sort_values('trial_conversion_rate', ascending=False)
axes[1].barh(df_rate['first_visit_page'], df_rate['trial_conversion_rate'], color=PALETTE[1])
axes[1].set_title('Trial Conversion Rate by First Page (%)', fontsize=12)
axes[1].set_xlabel('Conversion Rate (%)')
sns.despine(ax=axes[1])

fig.suptitle('Webpage Volume vs Conversion Rate', fontsize=14, y=1.02)
plt.tight_layout()
plt.savefig('../output/W4Q4_webpage_conversion_rate.png', bbox_inches='tight')
plt.show()
```

## Volume vs Conversion Rate — Opportunity Matrix

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(df['unique_visitors'], df['trial_conversion_rate'], s=80, color=PALETTE[0], alpha=0.8)
for _, row in df.iterrows():
    ax.annotate(row['first_visit_page'], (row['unique_visitors'], row['trial_conversion_rate']),
                textcoords='offset points', xytext=(6, 3), fontsize=9)
ax.set_title('Page Volume vs Conversion Rate — Opportunity Matrix', fontsize=13, pad=12)
ax.set_xlabel('Unique Visitors (volume)')
ax.set_ylabel('Trial Conversion Rate (%)')
sns.despine()
plt.tight_layout()
plt.savefig('../output/W4Q4_opportunity_matrix.png', bbox_inches='tight')
plt.show()
```
